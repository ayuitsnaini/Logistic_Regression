```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("../input/diabetes-uci-dataset/diabetes.csv")
```

```python
data.head()
```

```python

for i in data:
    if i != "Age":
        data[i] = LabelEncoder().fit_transform(data[i])
data.head()
```

```python
x = data.drop("class", axis=1)
y = data["class"]
```

```python

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=18)
lr.fit(x_train,y_train)
```

```python
y_head = lr.predict(x_test)
lr_acc = accuracy_score(y_test,y_head)
print("accuracy : ", lr_acc)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print("\n PRINT CONFUSION MATRIX DAN CLASSIFICATION REPORT")
```

```python
print("Confusion Matrix:")
cm = confusion_matrix(y_test, y_head)
print(cm)
```

```python
print("Classification report:")
cr = classification_report(y_test, y_head)
print(cr)
```
